```python
import pandas as pd
from pyphoon.db.data_extractor import DataExtractor
from pyphoon.db.pd_manager import PDManager
from os.path import exists, join
```

```python
orig_images_dir = '../../../../datasets/typhoon/wnp/image/'
corrected_dir = '../../../database/corrected'
db_dir = '../../../database/'

images = join(db_dir, 'images.pkl')
corrected = join(db_dir, 'corrected.pkl')
besttrack = join(db_dir, 'besttrack.pkl')
missing = join(db_dir, 'missing.pkl')

man = PDManager()
man.load_original_images(images)
man.load_corrected_images(corrected)
man.load_besttrack(besttrack)
man.load_missing_images_info(missing)
```

```python
extractor = DataExtractor(orig_images_dir, corrected_dir, man)
```

```python
triplets = extractor._get_one_hour_triplets(False)
```

```python
filenames = extractor.get_full_filenames(triplets, False)
```

```python
from os.path import abspath
```

```python
full_filenames = filenames.apply(abspath)
```

```python
triplet_files = triplets.applymap(lambda x: full_filenames[x])
```

```python
len(triplet_files)
```

```python
test_ratio = 0.05
```

```python
import random
random.seed(1)
```

```python
triplet_files['train'] = triplet_files.apply(lambda x: True if random.random() < test_ratio else False, axis=1)
```

```python
triplet_files.to_csv('triplets_training_set.csv')
```
